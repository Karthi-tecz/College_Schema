```python
import pandas as pd

try:
    df = pd.read_csv('COVID clinical trials.csv')
    display(df.head())
    print(df.shape)
except FileNotFoundError:
    print("Error: 'COVID clinical trials.csv' not found.")
```

```python
# Display data types of each column
print(df.info())

# Summarize descriptive statistics for numerical columns
print(df.describe())

# Identify unique values in categorical columns
for col in ['Status', 'Phases', 'Age']:
    print(f'\nUnique values in {col}:')
    print(df[col].unique())

# Check for missing values
print(df.isnull().sum())

# Examine the 'Start Date' column
print(df['Start Date'].head())
```

```python
# Identify categorical and numerical columns
categorical_cols = df.select_dtypes(include=['object']).columns
numerical_cols = df.select_dtypes(include=['number']).columns

# Impute missing values in categorical columns with "Unknown"
for col in categorical_cols:
    df[col].fillna("Unknown", inplace=True)

# Remove rows with missing values in numerical columns
df.dropna(subset=numerical_cols, inplace=True)

# Verify the effectiveness of the cleaning strategy
print(df.isnull().sum())
```

```python
# Identify categorical and numerical columns
categorical_cols = df.select_dtypes(include=['object']).columns
numerical_cols = df.select_dtypes(include=['number']).columns

# Impute missing values in categorical columns with "Unknown"
for col in categorical_cols:
    df[col] = df[col].fillna("Unknown")

# Remove rows with missing values in numerical columns
df.dropna(subset=numerical_cols, inplace=True)

# Verify the effectiveness of the cleaning strategy
print(df.isnull().sum())

# Create a summary table
summary_table = pd.DataFrame({
    'Column Name': df.columns,
    'Missing Values Before Cleaning': df.isnull().sum().values
})

# Add a cleaning method column
summary_table['Cleaning Method'] = 'Impute with "Unknown"'
for col in numerical_cols:
    summary_table.loc[summary_table['Column Name'] == col, 'Cleaning Method'] = 'Remove Rows'

display(summary_table)
```

```python
import matplotlib.pyplot as plt

# Univariate analysis for 'Status'
print(df['Status'].value_counts())
df['Status'].value_counts().plot(kind='bar', title='Status of Clinical Trials')
plt.show()

# Univariate analysis for 'Phases'
print(df['Phases'].value_counts())
df['Phases'].value_counts().plot(kind='bar', title='Distribution of Trial Phases')
plt.show()

# Univariate analysis for 'Age'
print(df['Age'].value_counts())
try:
    # Attempt numerical conversion (if possible)
    # Example:  df['Age_Numeric'] = pd.to_numeric(df['Age'], errors='coerce')
    # ... further analysis with numerical 'Age_Numeric' column ...
    df['Age'].value_counts().plot(kind='bar', title='Distribution of Age Groups')
    plt.show()

except (ValueError, TypeError):
    print("Unable to convert 'Age' column to numerical categories. The column contains non-numeric values which cannot be directly converted to numerical categories for plotting.")
    df['Age'].value_counts().plot(kind='bar', title='Distribution of Age Groups')
    plt.show()


print("\nSummary of Univariate Analysis:")
print("Status:")
print("Most frequent status categories provide insights into the current stage of most clinical trials.")

print("\nPhases:")
print("The distribution of trial phases indicates the concentration of trials in different development stages.")

print("\nAge:")
print("The distribution of age groups helps understand the target population of the trials. If 'Age' column is not numerical, consider manual conversion or alternative visualization methods.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Status vs. Phases
status_phase_cross_tab = pd.crosstab(df['Status'], df['Phases'])
print("Status vs. Phases Cross-Tabulation:\n", status_phase_cross_tab)

plt.figure(figsize=(12, 6))
sns.heatmap(status_phase_cross_tab, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Relationship between Status and Phases")
plt.show()


# Conditions vs. Outcome Measures
# Explode the 'Conditions' and 'Outcome Measures' columns
df_exploded_conditions = df.assign(Conditions=df['Conditions'].str.split('|')).explode('Conditions')
df_exploded_outcomes = df_exploded_conditions.assign(OutcomeMeasures=df_exploded_conditions['Outcome Measures'].str.split('|')).explode('OutcomeMeasures')

# Create a cross-tabulation of the exploded columns
condition_outcome_cross_tab = pd.crosstab(df_exploded_outcomes['Conditions'], df_exploded_outcomes['OutcomeMeasures'])

# Select top 10 conditions and outcomes for visualization
top_conditions = condition_outcome_cross_tab.sum(axis=1).nlargest(10).index
top_outcomes = condition_outcome_cross_tab.sum(axis=0).nlargest(10).index

filtered_cross_tab = condition_outcome_cross_tab.loc[top_conditions, top_outcomes]

plt.figure(figsize=(12, 6))
sns.heatmap(filtered_cross_tab, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Relationship between Top Conditions and Outcome Measures")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Convert 'Start Date' to datetime, handling errors
df['Start Date'] = pd.to_datetime(df['Start Date'], errors='coerce')

# Extract the year
df['Start Year'] = df['Start Date'].dt.year

# Group by 'Start Year' and count trials
trials_per_year = df.groupby('Start Year')['NCT Number'].count()

# Visualize the trend
plt.figure(figsize=(10, 6))
plt.plot(trials_per_year.index, trials_per_year.values, marker='o', linestyle='-')
plt.xlabel('Start Year')
plt.ylabel('Number of Trials')
plt.title('Trend of COVID-19 Clinical Trials Started Over the Years')
plt.grid(True)
plt.show()

# Describe the trend (optional, but recommended)
if trials_per_year.empty:
    print("No valid start dates found to analyze the trend")
else:
    print("Trend Description:")
    if trials_per_year.is_monotonic_increasing:
      print("The number of trials initiated shows an increasing trend over the years.")
    elif trials_per_year.is_monotonic_decreasing:
      print("The number of trials initiated shows an decreasing trend over the years.")
    else:
      print("The trend of trials initiated over the years is not consistently increasing or decreasing.")
```

```python
# prompt: Output and Visuals After running the code, you should observe: ● Bar charts showing the distribution of trial statuses, phases, and age groups. ● A time series plot illustrating the trend of trials over time.

import pandas as pd
import matplotlib.pyplot as plt
# Bar charts for distribution of trial statuses, phases, and age groups
print("Generating bar charts for distribution of trial statuses, phases, and age groups...")

# Status distribution
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='Status', order=df['Status'].value_counts().index, palette='viridis')
plt.title('Distribution of Clinical Trial Statuses')
plt.xlabel('Number of Trials')
plt.ylabel('Status')
plt.show()

# Phases distribution
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='Phases', order=df['Phases'].value_counts().index, palette='viridis')
plt.title('Distribution of Clinical Trial Phases')
plt.xlabel('Number of Trials')
plt.ylabel('Phases')
plt.show()

# Age group distribution
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='Age', order=df['Age'].value_counts().index, palette='viridis')
plt.title('Distribution of Clinical Trial Age Groups')
plt.xlabel('Number of Trials')
plt.ylabel('Age Group')
plt.show()

# Time series plot illustrating the trend of trials over time
print("\nGenerating time series plot for trial trends over time...")

# Ensure 'Start Date' is in datetime format (already done in the preceding code)
# df['Start Date'] = pd.to_datetime(df['Start Date'], errors='coerce')

# Extract the year from 'Start Date'
df['Start Year'] = df['Start Date'].dt.year

# Remove rows where 'Start Year' is NaN (due to conversion errors)
df_cleaned_dates = df.dropna(subset=['Start Year'])

# Group by 'Start Year' and count the number of trials
trials_over_time = df_cleaned_dates.groupby('Start Year').size().reset_index(name='Number of Trials')

# Sort by year for a proper time series plot
trials_over_time = trials_over_time.sort_values('Start Year')

plt.figure(figsize=(12, 7))
sns.lineplot(data=trials_over_time, x='Start Year', y='Number of Trials', marker='o')
plt.title('Trend of Clinical Trials Started Over Time')
plt.xlabel('Start Year')
plt.ylabel('Number of Trials')
plt.grid(True)
plt.show()

print("\nObservation Summary:")
print("Bar charts show the frequency distribution of trials across different statuses, phases, and age groups.")
print("The time series plot visualizes the number of new trials initiated each year, revealing the trend of trial activity over time.")
```
